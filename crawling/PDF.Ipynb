```python
from PyPDF2 import PdfReader, PdfWriter
import os

# 파일 경로 설정
input_path = '/Users/kimble/Downloads/The Cambridge Introduction to Literary Posthumanism_Joseph Tabbi 2024.pdf'
output_path = '/Users/kimble/Downloads/The Cambridge Introduction to Literary Posthumanism_Joseph Tabbi 2024_reversed.pdf'

# PDF 읽기
pdf_reader = PdfReader(input_path)
pdf_writer = PdfWriter()

# 페이지를 역순으로 새 PDF에 추가
for page in reversed(pdf_reader.pages):
    pdf_writer.add_page(page)

# 새 PDF 저장
with open(output_path, 'wb') as output_file:
    pdf_writer.write(output_file)

print("PDF 페이지 순서가 변경되어 저장되었습니다.")
```
